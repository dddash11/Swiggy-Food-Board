```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87 Safari/537.36'}
```

```python
food_list = []
```

```python
url_list = ["https://www.swiggy.com/restaurants/burger-king-opposite-to-west-side-shivaji-nagar-bangalore-5938",
           "https://www.swiggy.com/restaurants/the-belgian-waffle-co-v-v-puram-shankarapura-bangalore-197238",
           "https://www.swiggy.com/restaurants/polar-bear-vv-puram-shankarapura-bangalore-5893"]
```

```python
for url in url_list:
    r = requests.get(url, headers = header)
    soup = BeautifulSoup(r.text, 'html.parser')
    restaurant_name = soup.find("h1", {"class": "_3aqeL"}).text
    area = soup.find("div", {"class": "Gf2NS _2Y6HW"}).text
    ratings = float(soup.find("div", {"class": "_2l3H5"}).text)
    no_of_ratings_greater_than_equal_to = int(soup.find("div", {"class": "_1De48"}).span.text.split()[0][:-1])
    items = soup.find_all("div", {"class": "_2dS-v"})
    for j in range(len(items)):
        cat = items[j].find_all("div", {"class": "styles_detailsContainer__22vh8"})
        category = items[j].find("h2", {"class": ["M_o7R _27PKo", "M_o7R"]}).text
#         print("Category ", category)
        for i in range(len(cat)):
#             print("Item ", i)
            name = cat[i].find("h3", {"class": "styles_itemNameText__3ZmZZ"}).text
            price = float(cat[i].find("span", {"class": "rupee"}).text)
            desc = cat[i].find("div", {"class": "styles_itemDesc__3vhM0"})
            if desc != None:
                desc = desc.text
            else:
                desc = "No description available"
            bestseller_tag = cat[i].find("span", {"class": "styles_ribbon__3tZ21 styles_itemRibbon__353Fy"})
            if bestseller_tag != None:
                bestseller_tag = 1
            else:
                bestseller_tag = 0
                
            row = {
                'restaurant_name': restaurant_name,
                'area': area,
                'ratings': ratings,
                'no_of_ratings_greater_than_equal_to': no_of_ratings_greater_than_equal_to,
                'item_name': name,
                'item_category': category,
                'item_price': price,
                'item_description': desc,
                'bestseller_tag': bestseller_tag
                  }
            food_list.append(row)
#             print((restaurant_name, area, ratings, no_of_ratings_greater_than_equal_to, name, category, price, desc, bestseller_tag), "\n\n")
```

```python
df = pd.DataFrame(food_list)
```

```python
df.to_csv('swiggy_top_restaurant_foods_bengaluru.csv', index=False)
```
